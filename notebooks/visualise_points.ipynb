```python
import pickle
import numpy as np
import plotly.graph_objects as go

frame_path = "../ouster_frames_pkl/frame_000060.pkl"

with open(frame_path, "rb") as f:
    data = pickle.load(f)

pts_struct = data["points"]
```

```python
xyz = np.column_stack([
    pts_struct["x"],
    pts_struct["y"],
    pts_struct["z"]
]).astype(np.float32)
```

```python
print("XYZ shape:", xyz.shape)
```

```python

fig = go.Figure(
    data=[go.Scatter3d(
        x=xyz[:, 0],
        y=xyz[:, 1],
        z=xyz[:, 2],
        mode="markers",
        marker=dict(size=1),
    )]
)
fig.update_layout(
    scene=dict(
        xaxis_title="X",
        yaxis_title="Y",
        zaxis_title="Z",
    ),
    width=800,
    height=800,
)
fig.show()
```
