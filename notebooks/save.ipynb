```python
import pickle
import numpy as np
import open3d as o3d
import os

src_dir = "../ouster_frames_pkl"
dst_dir = "../ouster_pcd"
os.makedirs(dst_dir, exist_ok=True)

# читает один PKL и сохраняет PLY
def convert_pkl_to_ply(pkl_path, ply_path):
    with open(pkl_path, "rb") as f:
        data = pickle.load(f)

    pts = data["points"]

    xyz = np.column_stack([pts["x"], pts["y"], pts["z"]]).astype(np.float32)

    pcd = o3d.geometry.PointCloud()
    pcd.points = o3d.utility.Vector3dVector(xyz)

    o3d.io.write_point_cloud(ply_path, pcd)
    print("Saved:", ply_path)

# пример: сохранить один файл
convert_pkl_to_ply(
    f"{src_dir}/frame_000000.pkl",
    f"{dst_dir}/frame_000000.ply"
)
```
